# 3 Node line:

```julia
a1,a2,a3 = symbols("a_1, a_2, a_3", real=true)
α = [a1,a2,a3]
r = symbols("r", real=true)
order = 2
Dot(a,b) = sum( a .* b)
;
```

```julia
pow = [ r^i for i = 0:order]
```

```julia
eq = Dot(pow, α)
```

```julia
points = [-1,1,0]
```

```julia
a = [ subs(eq, r=>p) for p in points ]
```

```julia
A = [ diff(a,b) for a in a, b in α ]
```

```julia
A*α - a
```

```julia
N1 = transpose(pow)*inv(A) |> collect
```

```julia
N1r = [diff(N,r) for N in N1]
```

## Plot example

```julia
ue = [-1, 0.5, 1]
u= Dot( N1', ue )
```

```julia
plot(u, -1, 1, label="Shape Functions")
scatter!(points, ue, label="Control Points", xlabel="Natural coordinate r")
```
